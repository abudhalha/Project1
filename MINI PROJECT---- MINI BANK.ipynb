```python
class Bank:
    bankname = 'TIMES BANK'
    branch = 'KALPAKKAM'
    account_holders = []
    account_number_counter = 6698775302  # creating account number 

    def __init__(self, username,pannumber, mobile, address, pin_number, initial_deposit):
        self.username = username
        self.pannumber = pannumber
        self.mobile = mobile
        self.address = address
        self.pin_number = pin_number
        self.balance = initial_deposit
        self.account_number = Bank.account_number_counter  # creating account number
        Bank.account_number_counter += 1
        Bank.account_holders.append(self)
        print('=' * 66)
        print(f'Hello {self.username}! Your account with account number {self.account_number} has been created successfully.')
        print('=' * 66)
    
    def deposit(self, amount):
        self.balance += amount
        print(f'{amount} deposited successfully.')
        
    def withdraw_with_pin(self, pin, denomination):
        if pin == self.pin_number:
            if denomination <= self.balance:
                self.balance -= denomination
                print(f'Amount of {denomination} withdrawn successfully.')
            else:
                print('Insufficient funds.')
        else:
            print('Invalid PIN')
            
    def ministatement(self):
        print(f'Your account balance is {self.balance}.')
        
    def view_customer_records(self):
        print('\nCustomer Records:')
        for account in self.account_holders:
            print(f'Account Number: {account.account_number}, Username: {account.username}, Balance: {account.balance}')
            
# Creating an account
print('=' * 39)
print(f'Welcome to {Bank.bankname}, {Bank.branch}')
print('=' * 39)

print('*' * 39)
username = input('Enter Customer Name: ')
pannumber = input('Enter your PAN number')
mobile = input('Enter your 10 Digit Mobile Number: ')
address = input('Enter your Address: ')
pin_number = input('Please Enter A 4-digit PIN of your choice: ')
initial_deposit = float(input('Please Enter a Value to Deposit to Start an Account (min 500 and max 10000): '))

#b = Bank(username, pin_number, initial_deposit)
b = Bank(username, pannumber, mobile, address, pin_number, initial_deposit)

while True:
    print('\nSELECT YOUR CHOICE NUMBER FROM THE BELOW MENU:')
    print('1. Create an account\n2. Deposit\n3. Withdraw Money\n4. Mini statement\n5. View customer records\n6. Exit')
    option = int(input())
    
    if option == 1:
        username = input('Enter your name: ')
        pannumber = ('Enter your PAN number: ')
        mobile = input('Enter your mobile number: ')
        address = input('Enter your address: ')
        pin_number = input('Enter your PIN number: ')
        initial_deposit = float(input('Enter initial deposit amount: '))
        b = Bank(username, mobile, address, pin_number, initial_deposit)
        
    elif option == 2:
        amount = float(input('Enter deposit amount: '))
        b.deposit(amount)
        
    elif option == 3:
        pin = input('Enter PIN: ')
        denomination = float(input('Enter withdrawal amount: '))
        b.withdraw_with_pin(pin, denomination)
        
    elif option == 4:
        b.ministatement()
    
    elif option == 5:
        b.view_customer_records()
        
    elif option == 6:
        print('Thanks for using the Bank.')
        break
        
    else:
        print('Invalid option. Please select a valid option.')
        
```
